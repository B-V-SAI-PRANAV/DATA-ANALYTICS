```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

**Histogram: Visualizing Frequency Distribution**

```python
sns.histplot(df["price"])
```

```python
sns.histplot(df[df["price"] < 500]["price"])
```

```python
sns.histplot(df[df["number_of_reviews"] < 100]["number_of_reviews"])
```

**Distplot: Visualizing Probability Distribution**

```python
sns.distplot(df["price"])
```

**Box Plot: Identifying Outlier**s

```python
sns.boxplot(df[df["price"] < 500]["price"])
```

```python
sns.boxplot(df[df["reviews_per_month"] < 10]["reviews_per_month"])
```

**Violin Plot: Combining Distribution and Box Plot**

```python
df2 = sns.load_dataset("titanic")
sns.violinplot(df2["age"])
```
