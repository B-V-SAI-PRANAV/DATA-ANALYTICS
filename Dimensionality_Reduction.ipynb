```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/diabetes.csv')
data.head()
```

```python
data.describe()
```

```python
data=data.drop(columns=['Outcome'],axis=1)
```

```python
data.head()
```

```python
#converting the data into standard scaler form
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
data_scaled=scaler.fit_transform(data)
```

```python
data_scaled=pd.DataFrame(data_scaled)
data_scaled.describe()
```

```python
data_scaled.shape
```

```python
#finding the covariance matrix
covarianceMatrix=(data_scaled.T@ data_scaled)/767
covarianceMatrix
```

```python
#Evaluate the eigen values and the eigen vectors using the covariance matrix
eigen_values,eigen_vectors=np.linalg.eig(covarianceMatrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
#higher the eigen value , higher the information provided
PC1_data=data_scaled @ eigen_vectors[:,0]
PC2_data=data_scaled @ eigen_vectors[:,1]
PC3_data=data_scaled @ eigen_vectors[:,7]
PC4_data=data_scaled @ eigen_vectors[:,6]
```

**PCA implementation using sklearn library**

```python
from sklearn.decomposition import PCA
pca=PCA()
principalComponent=pca.fit_transform(data_scaled)
```

**how to know how many pca components could give us the optimal results ?**

```python
#to know the optimal number of pca components
pca.explained_variance_ratio_
```

```python
# now plot the graph from the obtained values , if at some point exceeds 90 percent the the number of
# components gives optimal results are obtained
import matplotlib.pyplot as plt
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('Explained variance ratio')
plt.savefig('pca.png')
plt.show()
```
